整數 (int)、浮點數 (float)

```python
x = 10
y = 2.4
print (x)
print (y)
```

布林值 (boolean)

```python
parse_ok = True
print(parse_ok)

parse_no_ok = False
print(parse_no_ok)
print(type(parse_no_ok))
```

字串 (string)

```python
name = "Oscar"
text = "False"
print(name)
print(text)
print(type(text))
```

列表 (list)

```python
my_list = [10, 5, 100]
print(my_list)
```

元組 (tuple)

```python
my_tuple = (1,5,10)
print(my_tuple)
```

集合 (set)

```python
my_set = {4, 6, 1}
print(my_set)
print(type(my_set))
```

字典 (dict)

```python
data = {'a':18, 'b':25}
print(data)
print(type(data))
```

```python
# 創建一個空的 dict
data = {}
print(data)
print(type(data))
```

```python
# 創建一個空的 set
data = set()
print(data)
```
