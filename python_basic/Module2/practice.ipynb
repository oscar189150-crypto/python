百貨公司周年慶 消費金額超過 2000 要折 200  

請撰寫一支程式  
計算客戶消費的結帳金額

```python
money = int(input("結帳金額："))

if (money>2000):
    discount = money//2000
    money = money-200*discount

print(money)
```
