### `json` 套件

| 用法 | 說明 |
|:--:|:--:|
| json.dump() |物件轉JSON並寫入文件|
| json.dumps(<list or dict>) |物件轉JSON字串|
| json.load() |讀取JSON文件|
| json.loads() |將JSON字串轉成Python物件|

```python
import json

fruit_dict = {
    "apple" : {
        "price" : 8, 
        "amount" : 10, 
        "color" : ["綠色", "紅色"], 
    }, 
    "orange" : {
        "price" : 20, 
        "amount" : 7, 
        "color" : ["橘色"], 
    }, 
    "tomato" : {
        "price" : 15, 
        "amount" : 1, 
        "color" : ["紅色"], 
    }, 
}

# json 格式 -> 只能是 list, dict
with open("fruit.json", "w") as f:
    f.write(json.dumps(fruit_dict))
```

```python
import json

fruit_dict = {
    "apple" : {
        "price" : 8, 
        "amount" : 10, 
        "color" : ["綠色", "紅色"], 
    }, 
    "orange" : {
        "price" : 20, 
        "amount" : 7, 
        "color" : ["橘色"], 
    }, 
    "tomato" : {
        "price" : 15, 
        "amount" : 1, 
        "color" : ["紅色"], 
    }, 
}

# json 格式 -> 只能是 list, dict
with open("fruit2.json", "w") as f:
    json.dump(fruit_dict, f , indent=4)
```

```python
import json

with open("fruit2.json", mode="r") as f:
    data = f.read()

print(type(data))
fruit_dict = json.loads(data)
print(type(fruit_dict))
print(fruit_dict)
```

```python
import json

with open("fruit2.json", mode="r") as f:
    data = json.load(f)

print(type(data))
print(data)
```
