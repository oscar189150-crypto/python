##### 將 data.txt 的資料分類成另外三個檔案，分別為 cc.txt 、 css.txt 、 c.txt  

```python
with open("data_need.txt", mode="r", ) as f:
    datalist = f.readlines()
data_list_without_n = []
print(datalist)
for data in datalist:
    data_list_without_n.append(data.strip("\n"))

print(data_list_without_n)

for text in data_list_without_n:
    if text[-2:] == "cc":
        with open("cc.txt", mode = "a") as f:
            f.write(text+"\n")
    elif text[-3:] == "css":
        with open("css.txt", mode = "a") as f:
            f.write(text+"\n")
    elif text[-1] == "c":
        with open("c.txt", mode = "a") as f:
            f.write(text+"\n")



```
