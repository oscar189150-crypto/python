條件語句 if-elif-else

```python
'''
條件判斷
if ...
if ... else ...
if ... elif ... else ...

'''
```

##### if

```python
a = 5
b = 7

if (b>a):
    print("true")
```

##### if...else...

```python
a = 5
b = 7

if (a>b):
    print("True")
else:
    print("False")
```

##### if...elif... (else)

```python
fruit = "bannan"
if fruit == "apple" :
    print("this is appple")
elif fruit == "banana" :
    print("this is banana")
elif fruit == "bannan" :
    print("這是捷運^^")
```
