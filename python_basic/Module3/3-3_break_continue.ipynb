##### break

```python
'''
當偵測到字母 t 時，就會強制結束迴圈
'''
text = "prectice"
for char in text:
    if char == "e":
        break
    print(char, end=", ")

```

##### continue

```python
'''
當偵測到字母 t 時，
會跳過本次迴圈剩下的程式碼 print(char)，
但不會結束迴圈，仍然會進入下一圈繼續執行
'''
text = "prectice"
for char in text:
    if char == "e":
        continue
    print(char, end=", ")


```

```python
x = 0

for i in range(1, 101):
    x += i
    print(x, end=", ")
    if x > 500:
        break
```
