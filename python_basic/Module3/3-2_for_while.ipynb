| 用法 | 說明 |
|:---:|:---:|
| `range(m)` | 走訪 0 到 m-1 的數字 |
| `range(n, m)` | 走訪 n 到 m-1 的數字  |
| `range(n, m, step)` | 以step為間距，走訪 n 到 m-1 的數字 |

##### for

```python
#for <變數> in <資料>:
for something in 'ThisIsAWord':#直接給予資料
    print(something, end=", ")
```

```python
data = [1, 2, 3, 4, 5, 6, 7]
for num in data:#先把資料存在別的變數
    if num > 3:
        print(num, end=", ")

```

```python
# range(m)<<遍歷 0 ~ m-1的數字
for i in range(2):
    print(i, end=", ")
```

```python
# range(n, m-1) 走訪n ~ m-1的數字
for i in range(1, 11):#從1 ~ 10
    print(i, end=", ")
```

```python
# range(n, m-1, step) step = 每次加多少
for i in range(1, 11, 2):
    print(i, end=(', '))

```

### 小試身手 for  
請產出 a 開頭的 n 個偶數(含 a 本身)  
Ex:  
a = 5, n = 3  
Output : 6 8 10

a = 4, n = 5  
Output : 4 6 8 10 12

```python
star = int(input("請輸入起始數字："))
time = int(input("幾個數字？"))

for i in range(star, star+2*time, 2):
    if star%2 == 0:
        print(i, end=", ")
    elif star%2 == 1:
        print(i+1, end=", ")
```

```python
x = int(input("你想要幾層三角形？"))

for i in range(x):
    print(" "*(x-1-i) + ("* "*(i+1)))
print("")
for i in range(x):
    print(" "*(i) + ("* "*(x-i)))
```
