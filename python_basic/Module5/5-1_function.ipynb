```python
# 建立 function 2
def get_name():
    name = "Oscar"
    return name

result = get_name()
print(result)
```

```python
# 建立 function 3
def greeting(name):
    print(f"Hello, {name}")

greeting("Alex")
```

```python
# 建立 function 4
def greeting(name, greeting = "Hello"):
    print(f"{greeting}, {name}")
greeting("amy")
greeting("amy", "Good Morning")
greeting(greeting = "Good Morning", name = "amy")
```

```python
def add (*arg, **kwargs):
    print(kwargs)
    total = 0
    for i in arg:
        total += i
    return total

total = add(3, 5, 11, 20, 38, unit="元")
print(total)
```
