### lambda 匿名函式  
用法  

`function_name = lambda arguments: expression`

````python
# 建立函式
def add(x, y):
    return x + y

result = add(10, 5)
print(result)
````

````python
# 基本用法
add2 = lambda x, y: x + y
result = add2(7, 8)
print(result)
````

````python
# 條件判斷
abs = lambda x : x if x > 0 else -x
result = abs(-8)
print(result)
````

### 小試身手  
請建立一個 list ，並在該 list 內產生多個 `[1, 2, x]` 的 list (x 的值要遞增)  
x 到 10 即可  
Ex :  
```Python
[ [1, 2, 1], [1, 2, 2], [1, 2, 3]...[1, 2, 10] ]
```

````python
ans = []
for i in range(1, 11):
    ans.append([1, 2, i])
print(ans)
````

````python
ans1 = [[1, 2, i] for i in range(1, 11)]
print(ans1)
````

##### `sort()` 方法

````python
ans = []
for i in range(1, 11):
    ans.append([1, 2, i])
print(ans)

#預設是由小排到大

ans.sort(reverse=True)#由大排到小
print(ans)
````

每三個 list 中間的值要改成 5  
Ex:  
```python
[ [1, 2, 1], [1, 2, 2], [1, 5, 3], [1, 2, 4], [1, 2, 5], [1, 5, 6]...]
```

````python

ans = []
for i in range(1, 11):
    if i % 3 == 0:
        ans.append([1, 5, i])
    else:
        ans.append([1, 2, i])
print(ans)

ans.sort(reverse=True, key= lambda x: x[2])
print(ans)
````
