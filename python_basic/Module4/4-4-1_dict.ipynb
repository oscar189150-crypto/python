```python
# 初始化 dict
fruit_price = {"apple" : 50, "banana" : 30, "tomato" : 80}
print(fruit_price)
```

```python
#當資料多的時候 建議的寫法
fruit_price = {
    "apple" : 50, 
    "banana" : 30, 
    "tomato" : 100, 
    "orange" : 200
}
```

```python
# 印出 蘋果的價格 -> 取得某key的value

print(fruit_price["apple"])
```

```python
# 修改 橘子的價格
fruit_price["orange"] = 80
print(fruit_price)
```

```python
# 刪除 番茄
fruit_price.pop("tomato")
print(fruit_price)
```

```python
# 建立一個 value 含有 dict 的 dict
user_info = {
    "name" : "ula", 
    "age" : 20, 
    "number" : {
        "home" : "034567890", 
        "cellphone" : "0987654321", 
        "abordphone" : "+1 234 567 890", 
    }, 
    "email" : "asdf@gmail.com", 
    "address" : "桃園市桃園區桃園火車站" 
}
```

```python
# 取得 nickname
print(user_info)
print(user_info["address"])

```

```python
# 取得所有 phone_number
print(user_info["number"]["home"])
```

```python
'''
改寫 fruit_dict 每個水果需要 price, quantity/amount, color
apple 有 綠色 和 紅色
orange 有 橘色
tomato 有 紅色
'''
fruit_dict = {
    "apple" : {
        "price" : 8, 
        "amount" : 10, 
        "color" : ["綠色", "紅色"], 
    }, 
    "orange" : {
        "price" : 20, 
        "amount" : 7, 
        "color" : ["橘色"], 
    }, 
    "tomato" : {
        "price" : 15, 
        "amount" : 1, 
        "color" : ["紅色"], 
    }, 
}

```

##### dict 的合併

```python
# dict 的合併
# <dict>.update()
fruit_dict2 = {
    "kiwi" : {
        "price" : 50, 
        "amount" : 200, 
        "color" : ["綠色", "金色"], 
    }, 
}

fruit_dict.update(fruit_dict2)
print(fruit_dict)
```

##### 另外的 print 方式 `pprint` (pretty-print) 

```python
# 從 Module 引入 function
from pprint import pprint
pprint(fruit_dict, sort_dicts=False)#不要亂排序
```

```python
'''
新增顏色
apple 有新的品種 顏色要有黃色
tomato 有綠色
'''
new_color = "黃色"
if new_color not in fruit_dict["apple"]["color"]:
    fruit_dict["apple"]["color"].append("黃色")
new_color = "綠色"
if new_color not in fruit_dict["tomato"]["color"]:
    fruit_dict["tomato"]["color"].append("綠色")
pprint(fruit_dict, sort_dicts=False)
```
