```python
# f-string (又稱作 formatted string literals, version >= 3.6)

```

```python
# 靠左對齊、靠右對齊
age = 20
name = "Ula"
msg = f"name:[{name:<10}], age:[{age:>10}]"
print(msg)
#置中
msg = f"name:[{name:^10}], age:[{age:^10}]"
print(msg)
```

```python
# 靠左對齊、靠右對齊，字數不足，補「ㄏ」
# (沒寫補什麼，預設空格)
msg = f"name:[{name:ㄏ<10}], age:[{age:ㄏ>10}]"
print(msg)
msg = f"name:[{name:X^10}], age:[{age:X^10}]"
print(msg)
```

```python
# 置中對齊
msg = f"name:[{name:^10}], age:[{age:^10}]"
print(msg)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
msg = f"name:[{name:^10.2}], age:[{age:^10}]"
print(msg)

```

```python
ID = "F123456789"
msg = f"[{ID:*<10.3}]"
print(msg)
```
