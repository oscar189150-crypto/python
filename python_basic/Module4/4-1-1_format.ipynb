|語法|說明|
|:---:|:---:|
|%s|插入 str字串 的值到字串中|
|%f|插入 浮點數 的值到字串中|
|%d|插入 整數 的值到字串中|
|%%|輸出 百分比 (類似跳脫字元)|

```python
# 多組文字
data1 = "hello"
data2 = "world"

msg = "%s, %s!" %(data1, data2)
print(msg)
```

```python
# 整數
age = 70
msg = "I am %d years old." %(age)
print(msg)
```

```python
# 文字與整數
name = "Jesson"
age = 12
msg = "%s is %d years old." %(name, age)
print(msg)
```

```python
# 指定寬度 (維持 10 個字元長度，預設向右看齊)
msg = "[%10s]" % "Hello"
print(msg)
```

```python
# 靠左對齊 (維持 10 個字元長度，向左對齊)
msg = "[%-10s]" % "Hello"
print(msg)
```

```python
# 指定浮點數位數
num = 3.1415926535
msg = "%5f" %num
print(msg)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
msg = "%.3s" % "Hello"
print(msg)
```

```python
# 空白補 0
msg = "%06.2f" % 3.14
print(msg)
```
