```python
fruit_dict = {
    "apple" : {
        "price" : 8000, 
        "amount" : 10, 
        "color" : ["綠色", "紅色"], 
    }, 
    "orange" : {
        "price" : 20, 
        "amount" : 7, 
        "color" : ["橘色"], 
    }, 
    "tomato" : {
        "price" : 1, 
        "amount" : 1, 
        "color" : ["紅色"], 
    }, 
}
max_fruit , max_price = "", 0
for key, value in fruit_dict.items():
    if value["price"] > max_price:
        max_fruit = key
        max_price = value["price"]
print(f"最貴的水果是: {max_fruit}, 他賣{max_price}元")
```

```python
# 資料處理
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}]
```

```python
# 將資料整理成 {'apple': ['red', 'green'], 'grape': ['purple', 'green'], 'kiwi': ['yellow', 'green']}

after_fruit_color_list = {}
for i in fruit_color_list:# i = 字典dict
    # print(i)
    for key, value in i.items():# key = 字串str, value = 字串str
        if key in after_fruit_color_list:
            after_fruit_color_list[key].append(value)
        elif key not in after_fruit_color_list:
            after_fruit_color_list[key] = [value]

print(after_fruit_color_list)
```

##### `defaultdict(<type>)`

```python
# from 模組 import 函式/class
# 當訪問的鍵不存在時，可以自動創建鍵和預設值，而不是拋出KeyError。
from collections import defaultdict
data = defaultdict(int)
# 以int型態的預設值 0 來初始化
print(data["apple"])
print(data)
```

```python
result = defaultdict(list)

print(result["apple"])
print(result)
```

```python
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}]
result = defaultdict(list)

for fruit_color in fruit_color_list:
    for fruit, color in fruit_color.items():
        if color not in result[fruit]:
            result[fruit].append(color)
print(result)
```
