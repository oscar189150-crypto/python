```python
# 初始化一個 list
num_list = [1, 2, 3]
print(num_list)
```

```python
# 新增元素在 list 尾端
#<list>.append(<data>)
num_list = [1, 2, 3]
num_list.append(5)
print(num_list)
```

```python
# 修改索引位置的元素
#<list>[<位置>] = <data>
num_list = [10, 20, 30]
num_list[2] = 99
print(num_list)
```

```python
# 刪除索引位置的元素
#<list>.pop(<位置>)
num_list = [10, 20, 30]
num_list.pop(0)
print(num_list)
```

```python
# 新增元素 9 在指定索引 1，其餘元素往後移
#<list>.insert(<位置>, <值>)
num_list = [10, 20, 30]
num_list.insert(1, 99)
print(num_list)
```

```python
# 移除指定的值
#<list>.remove(<指定的值>)
num_list = [10, 20, 30, 20, 70, 90, 30, 30, 30]
num_list.remove(30)
print(num_list)

while 30 in num_list:
    num_list.remove(30)
print(num_list)
```

```python
# 計算 list 內的某個值 共有幾個
#<list>.count(<值>)
# list裡面有多少元素
#len(<list>)
num_list = [10, 20, 30, 20, 70, 90, 30, 30, 30]
print(f"資料裡面有{num_list.count(30)}個30")
print(f"資料裡面有{len(num_list)}個元素")
```

### list 搭配 for 迴圈

```python
# 建立一個 1~10 的 list
nums = []
for i in range(1, 11):
    nums.append(i)
print(nums)

nums = []
for i in range(1, 11):
    nums.append(i**2)
print(nums)
```

```python
# enumerate() 同時取得 index 與 value
# enumerate() 這東西會回傳兩個變數
# 回傳的內容是 1.列表裡面的位置 2.那個位置的內容
nums = []
for i in range(1, 11):
    nums.append(i**2)
print(nums)
# 所以底下這邊用兩個變數去接他
for index, value in enumerate(nums):
    print(f"index:{index} 值:{value}")

for index, value in enumerate(nums):
    if value == 49:
        print(f"49在位置:{index}")
```

```python
'''
建立一個排隊的系統
queue = ["John", "Andy", "Kevin"]
1. 新的客人進來 Sandy
2. "John" 完成交易 要離開
3. 有 VIP 客人來 Boss 要排第一個 


預期輸出:
["Boss", "Andy", "Kevin", "Sandy"]
'''
queue = ["John", "Andy", "Kevin"]
queue.append("Sandy")
print(f"1.{queue}")
queue.pop(0)
print(f"2.{queue}")
queue.insert(0, "Boss")
print(f"3.{queue}")
```

### List comprehension 列表推導式
`new_list = [expression for item in iterable [if condition]]`  
  
`list的名稱 = [表達式 for 元素 in 迭代對象 [if 判斷式(...的條件)]]`

```python
num = []
for i in range(1, 21):
    num.append(i)

print(num)

num = [i for i in range(1, 21) if i % 5 ==0]
print(num)
```

```python
# 列出 n 以內的質數
n = 100
list = []

for i in range(2, n+1):
    list.append(i)
    for j in range(2, int(i**0.5)+1):
        if i % j == 0:
            list.remove(i)
            break

print(list)
```

```python
#合併兩個list
#直接改變原先的list
nums1 = [1, 2, 3]
nums2 = [7, 8, 9]
nums1.extend(nums2)
print(nums1)
```

```python
#兩個list也可以直接相加
nums1 = [1, 2, 3]
nums2 = [7, 8, 9]
#但是會需要一個新的變數來承接它
all_nums = nums1 + nums2
print(all_nums)
```

```python
#合併list錯誤用法
nums1 = [1, 2, 3]
nums2 = [7, 8, 9]
nums1.append(nums2)
print(nums1)
#建議不要把不同型別的資料放在同一個list
#同一個list只建議放一種型別的資料
```
