```python
# 替換字串
'''
str.replace(<old>, <new>)
將 string 中的 str1 替換成 str2
'''
name = "Alex"
name1 = name.replace("ex", "lison")
print(name1)
```

```python
# 去除兩側空格
'''
data.strip()
去除字串 string 左、右兩邊的空格
'''
data = "xxxMisakixxx"
data1 = data.strip("x")#去除兩側所有的 x
print(data1)

data2 = data.lstrip("x")#只去除左邊的 x
print(data2)

data3 = data.rstrip("x")#只去除右邊的 x
print(data3)

```

```python
# 字串小寫變大寫
'''
str.lower()
將字串 string 裡的字母全部改成小寫
'''
data = "BANANA"
data1 = data.lower()
print(data1)

'''
str.upper()
將字串 string 裡的字母全部改成大寫
'''
```

```python
# 字串分割成 list
# str.split(<用來分隔的東西>)
greeting = "hello world! hello John"
greeting1 = greeting.split(" ")
greeting2 = greeting.split("!")
print(greeting1)
print(greeting2)



#字串合併 str.join(<可迭代物件>)
greeting = "_".join(data)
print(greeting)

greeting = "___".join(("AP", "BD"))
print(greeting)
```
