```python
# tuple 初始化: 第一種
my_tuple = ("Big", "Data", "Is", "Good")
print(my_tuple)
```

```python
# tuple 初始化: 第二種
my_tuple2 = "Small", "Data", "Is", "Good"
print(my_tuple2)
```

```python
# 透過指定索引輸出值
print(my_tuple[1])
print(my_tuple2[1])
```

```python
# 複數變數修改值
data = ("AIPE", "BDSE")
print(data)

a, b = b, a
print(a)
print(b)
```

```python
# tuple 不可以修改指定索引的值
data = ("AIPE", "BDSE")
print(data)

a, b = data
print(a)
print(b)
```

```python
# (list)可以修改指定索引的值
nums = (1, 2, 3)
nums = list(nums)
print(nums)
nums[1] = 20
print(nums)
```

```python
# 用 for 迴圈逐一輸出資料
for value in nums:
    print(value)
```

```python
# 用 len 計算 tuple 資料個數
print(len(nums))
```
